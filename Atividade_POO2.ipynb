```python
from abc import ABC, abstractmethod

class usuario(ABC):
  def __init__(self, nome):
    self.nome = nome

  @abstractmethod
  def informacoes(self):
    pass

class aluno(usuario):
  def __init__(self, nome, curso, turma):
    super().__init__(nome)
    self._matricula = []
    self.curso = curso
    self.turma = turma

  def matricular(self, curso, turma):
    nova_matricula = matricula(self, turma)
    if turma.vagas_dispo() > 0:
      self._matricula.append(nova_matricula)
      turma.adicionar(self)
      print("Mátricula realizada com sucesso!")
    else:
      print("Não há vagas nessa turma.")

  def informacoes(self):
    print(f"Nome: {self.nome}\nMatricula: {self.matricula}\nCurso: {self.curso}\nTurma: {self.turma}")

class coordenador(usuario):
  def __init__(self, nome):
    super().__init__(nome)
    self.cursos = []

  def cadastrar(self, nome, preco, carga):
    novo_curso = curso(nome, preco, carga)
    self.cursos.append(novo_curso)
    print("Curso cadastrado com sucesso! ")

  def informacoes(self):
    print(f"Coordenador: {self.nome}\n Curso: {curso.nome}\n")

class professor(usuario):
  def __init__(self, nome, materia):
    super().__init__(nome)
    self.materia = materia

  def informacoes(self):
    print(f"Professor: {self.nome}\n Matéria de ensino: {self.materia}\n")

class curso():
  def __init__(self, nome, preco, carga):
    self.nome = nome
    self.preco = preco
    self.carga = carga
    self.turmas = []

  def adc_turma(self, turma):
    self.turmas.append(turma)

  def informacoes(self):
    print(f"Curso: {self.nome}\nPreço: {self.preco}\nCarga horária: {self.carga}")

class turma():
  def __init__(self, data, instrutor, vagas):
    self.data = data
    self.instrutor = instrutor
    self.vagas = vagas
    self.alunos = []

  def vagas_dispo(self):
    contador = 0
    for aluno in self.alunos:
      contador += 1
    return self.vagas - contador

  def adicionar(self, aluno):
    self.alunos.append(aluno)

  def informacoes(self):
    print(f"Data: {self.data}\nInstrutor: {self.instrutor}\nVagas: {self.vagas}")

class matricula():
  def __init__(self, aluno, turma):
    self.aluno = aluno
    self.turma = turma
    self.pagamento = None

  def pagar(self, pagamento):
    self.pagamento = pagamento
    pagamento.pagar()
    print("Pagamento realizado com sucesso! ")


class pagamento(ABC):
  def __init__(self, valor):
    self.valor = valor

  @abstractmethod
  def pagar(self):
    pass

class boleto(pagamento):
  def __init__(self, valor, data):
    super().__init__(valor)
    self._data = data

  def pagar(self):
    print("Pagamento realizado com suceso!")

class pix(pagamento):
  def __init__(self, valor, chave):
    super().__init__(valor)
    self.__chave = chave

  def pagar(self):
    print("Pagamento realizado com suceso!")

class cartao(pagamento):
  def __init__(self, valor, numero, validade, cvv):
    super().__init__(valor)
    self._numero = numero
    self._validade = validade
    self.__cvv = cvv

  def pagar(self):
    print("Pagamento realizado com suceso!")
```

```python

cursos = []
alunos = []

while True:
  print("---- CURSOS XYZ ----")
  print("1) Matricular aluno")
  print("2) Cadastrar curso")
  print("3) Cadastrar turma")
  print("4) Efetuar pagamento")
  print("5) Listar cursos e turmas disponíveis")
  print("6) Sair")
  opcao = int(input("Digite a opção desejada: "))

  if opcao == 1:
    if not cursos:
      print("Nenhum curso cadastrado no momento. ")
    else:
      nm = input("Digite o nome do aluno: ")
      curso_aluno = input("Digite o nome do curso para matricular: ")
      turma_aluno = input("Digite a data de início e fim da turma para matricular(MM/DD MM/DD): ")
      novo_aluno = aluno(nm, curso_aluno, turma_aluno)
      alunos.append(novo_aluno)
      print("Aluno cadastrado com sucesso! ")

  elif opcao == 2:
    nome = input("Digite o nome do curso: ")
    preco = float(input("Digite o preço do curso (apenas valor): "))
    carga = int(input("Digite a carga horária do curso (apenas a hora): "))
    novo_curso = curso(nome, preco, carga)
    cursos.append(novo_curso)
    print("Curso cadastrado com sucesso! ")

  elif opcao == 3:
    if not cursos:
      print("Nenhum curso cadastrado no momento. ")
    else:
      nome_curso = input("Digite o nome do curso para adicionar a turma: ")
      found_curso = None
      for c in cursos:
        if c.nome == nome_curso:
          found_curso = c
          break

      if found_curso:
        prof = input("Digite o nome do professor: ")
        data = input("Digite a data de início e de término da turma (MM/DD MM/DD): ")
        vagas = int(input("Digite a quantidade de vagas: "))
        nova_turma = turma(data, prof, vagas)
        found_curso.adc_turma(nova_turma)
        print("Turma cadastrada com sucesso! ")
      else:
        print("Curso não encontrado. ")


  elif opcao == 4:
    if not alunos:
      print("Nenhum aluno matriculado no momento. ")
    else:
      nome_aluno_pagamento = input("Digite o nome do aluno para efetuar o pagamento: ")
      found_aluno = None
      for a in alunos:
        if a.nome == nome_aluno_pagamento:
          found_aluno = a
          break

      if found_aluno:
        if not found_aluno.matricula:
          print("Este aluno não possui matrículas para pagar. ")
        else:
          opcao_pagamento = input("Digite o método de pagamento(pix/cartao/boleto): ")
          matricula_a_pagar = found_aluno.matricula[0]
          if opcao_pagamento == "pix":
            valor = float(input("Digite o valor a ser pago: "))
            chave = input("Digite a chave pix: ")
            pagamento_realizado = pix(valor, chave)
            matricula_a_pagar.pagar(pagamento_realizado)
            print("Pagamento realizado com sucesso! ")
          elif opcao_pagamento == "cartao":
            valor = float(input("Digite o valor a ser pago: "))
            numero = input("Digite o número do cartão: ")
            validade = input("Digite a data de validade do cartão: ")
            cvv = input("Digite o código de segurança do cartão: ")
            pagamento_realizado = cartao(valor, numero, validade, cvv)
            matricula_a_pagar.pagar(pagamento_realizado)
            print("Pagamento realizado com sucesso! ")
          elif opcao_pagamento == "boleto":
            valor = float(input("Digite o valor a ser pago: "))
            data_vencimento = input("Digite a data de vencimento do boleto: ")
            pagamento_realizado = boleto(valor, data_vencimento)
            matricula_a_pagar.pagar(pagamento_realizado)
            print("Pagamento realizado com sucesso! ")
          else:
            print("Opção inválida. ")
      else:
        print("Aluno não encontrado. ")


  elif opcao == 5:
    if not cursos:
      print("Nenhum curso cadastrado no momento. ")
    else:
      for c in cursos:
        print(f"Curso: {c.nome}")
        for t in c.turmas:
          print(f"  Turma: {t.data}")
          print(f"  Instrutor: {t.instrutor}")
          print(f"  Vagas disponíveis: {t.vagas_dispo()}")
        print()

  elif opcao == 6:
    print("Saindo do programa... ")
    break
  else:
    print("Opção inválida. ")
```
